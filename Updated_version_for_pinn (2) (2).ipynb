# Model Creation 

```python
class PINN(Model):
    def __init__(self):
        super(PINN, self).__init__()
        self.dense1 = layers.Dense(128, activation='relu')
        self.dense2 = layers.Dense(64, activation='relu')
        self.dense3 = layers.Dense(32, activation='relu')
        self.dense4 = layers.Dense(16, activation='relu')
        self.dense5 = layers.Dense(1)

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        x = self.dense3(x)
        x = self.dense4(x)
        return self.dense5(x)
```

## Physics Loss ( Heat Equation along with Data loss )

```python
def custom_loss(y_true, y_pred, X_train):
    # Data loss (MSE between predicted and true temperatures)
    data_loss = tf.reduce_mean(tf.square(y_true - y_pred))
    
    # Physics-informed regularization term (2D heat equation)
    alpha = 1.0  # Thermal diffusivity
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(X_train)
        y_pred_with_grad = model(X_train)
        
        # Compute first derivatives
        dT_dx = tape.gradient(y_pred_with_grad, X_train)[:, 0]  
        dT_dz = tape.gradient(y_pred_with_grad, X_train)[:, 2]  
        
        # Compute second derivatives
        d2T_dx2 = tape.gradient(dT_dx, X_train)[:, 0]  
        d2T_dz2 = tape.gradient(dT_dz, X_train)[:, 2]  
    
    # Laplacian (sum of second derivatives)
    laplacian = d2T_dx2 + d2T_dz2
    
    # Physics loss (enforce Laplacian = 0)
    physics_loss = tf.reduce_mean(tf.square(laplacian))
    
    # Total loss (data loss + physics loss)
    total_loss = data_loss + physics_loss
    return total_loss
```

## Model Training 

```python
model = PINN()
model.compile(optimizer='adam', loss=lambda y_true, y_pred: custom_loss(y_true, y_pred, tf.convert_to_tensor(X_train_1, dtype=tf.float32)))
```

```python

# Example data
x_coords = np.array([-17.68, -12.62, -7.57, -2.52, 2.53, 7.57, 12.62, 17.67])
z_coords = np.array([-12.61, -12.61, -12.61, -12.61, -12.61, -12.61, -12.61, -12.61])

# Actual temperatures
actual_temperatures = np.array([430, 415.5, 401.08, 372.18, 357.7, 328.8, 314.4, 300])

# Predicted temperatures
predicted_temperatures = np.array([395.81247, 383.06427, 369.29327, 355.36563, 341.43802, 327.53842, 314.1271, 300.71588])

# Add small noise to z_coords to make the data 2D
np.random.seed(42)  # For reproducibility
z_coords += np.random.normal(0, 1e-6, size=z_coords.shape)

# Create a grid for plotting
x_grid, z_grid = np.meshgrid(np.linspace(x_coords.min(), x_coords.max(), 100),
                             np.linspace(z_coords.min(), z_coords.max(), 100))

# Interpolate actual and predicted temperatures onto the grid
actual_grid = griddata((x_coords, z_coords), actual_temperatures, (x_grid, z_grid), method='linear')
predicted_grid = griddata((x_coords, z_coords), predicted_temperatures, (x_grid, z_grid), method='linear')

# Calculate the difference between actual and predicted temperatures
difference_grid = actual_grid - predicted_grid

# Create a figure with three subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 8))

# Define a common scale for both plots
vmin = actual_temperatures.min()  # Minimum temperature
vmax = actual_temperatures.max()  # Maximum temperature

# Plot actual temperatures (left subplot)
heatmap_actual = ax1.contourf(x_grid, z_grid, actual_grid, levels=50, cmap='Reds', vmin=vmin, vmax=vmax)
cbar_actual = plt.colorbar(heatmap_actual, ax=ax1, label='Temperature')
ax1.scatter(x_coords, z_coords, c='blue', label='Nodes')
ax1.set_xlabel('X Coordinate', fontsize=12)
ax1.set_ylabel('Z Coordinate', fontsize=12)
ax1.set_title('Actual Temperature Distribution', fontsize=14)
ax1.legend()

# Plot predicted temperatures (middle subplot)
heatmap_predicted = ax2.contourf(x_grid, z_grid, predicted_grid, levels=50, cmap='Reds', vmin=vmin, vmax=vmax)
cbar_predicted = plt.colorbar(heatmap_predicted, ax=ax2, label='Temperature')
ax2.scatter(x_coords, z_coords, c='blue', label='Nodes')
ax2.set_xlabel('X Coordinate', fontsize=12)
ax2.set_ylabel('Z Coordinate', fontsize=12)
ax2.set_title('Predicted Temperature Distribution', fontsize=14)
ax2.legend()

# Plot temperature difference (right subplot)
heatmap_difference = ax3.contourf(x_grid, z_grid, difference_grid, levels=50, cmap='coolwarm')
cbar_difference = plt.colorbar(heatmap_difference, ax=ax3, label='Temperature Difference')
ax3.scatter(x_coords, z_coords, c='blue', label='Nodes')
ax3.set_xlabel('X Coordinate', fontsize=12)
ax3.set_ylabel('Z Coordinate', fontsize=12)
ax3.set_title('Temperature Difference (Actual - Predicted)', fontsize=14)
ax3.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
